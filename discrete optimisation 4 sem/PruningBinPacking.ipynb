```python
total = int(input()) # total_knapsack_volume
number = int(input()) # number_of_items
Bin = [] # [item_volume, item_value]
import numpy as np
for i in np.arange(number):
    Bin.append([int(j) for j in input().split()])
Bin = [Bin[i] for i in np.arange(number) if Bin[i][0] < total]
number = len(Bin)

def sort_key(i): # sort Bin in price/volume order
    return -i[1]/i[0]

Bin.sort(key = sort_key)
Bin = np.asarray(Bin)

r = [0 for i in range(0, number)]
vol1 = 0
price1 = 0
for i in range(0, number):
    vol1 += Bin[i][0]
    if vol1 <= total:
        price1 += Bin[i][1]
        r[i] = 1
    else:
        vol1 -= Bin[i][0]
        
def greed(total_1, number_1): #return lower bound
    if number_1 == number:
        return 0
    vol = 0
    price = 0
    for i in range(number_1, number):
        vol += Bin[i][0]
        if vol <= total_1:
            price += Bin[i][1]
        else:
            vol -= Bin[i][0]
    return float(price)
               
def FakeLp(total_3, number_3):
    if number_3 == number:
        return 0
    vol = 0
    price = 0
    for i in range(number_3, number):
        vol += Bin[i][0]
        if vol <= total_3:
            price += Bin[i][1]
        else:
            vol -= Bin[i][0]
            price += Bin[i][1]*(total_3 - vol)/Bin[i][0]
            break
    return float(price)

def attempt(total_c, number_c, cost):
    global Min
    ch = 1
    var_xi = 0
    while ch == 1:
        while number_c != number and Bin[number_c][0] > total_c:
            number_c += 1
        if number_c == number:
            return cost
        lp =  FakeLp(total_c, number_c + 1)
        if Min > cost + lp:
            cost += Bin[number_c][1]                                                                          
            total_c -= Bin[number_c][0]
            number_c += 1
            continue
        gr = greed(total_c, number_c)
        if Min <= gr + cost:
            Min = gr + cost
            #r[number_c] = 1
        ch = 1 if gr >= lp else 0
        if ch:
            #r[number_c] = 1
            cost += Bin[number_c][1]                                                                          
            total_c -= Bin[number_c][0]
            number_c += 1
        else:
            var_xi = attempt(total_c - Bin[number_c][0], number_c + 1, cost + Bin[number_c][1])
            if var_xi > Min:
                Min = var_xi
                #r[number_c] = 1
            else:
                if r[number_c] == 0:
                    number_c += 1
                    continue
            if float(var_xi) >= cost + lp:
                return var_xi
            else:
                gr = greed(total_c, number_c + 1)
                if Min < gr + cost:
                    Min = gr + cost
                    #r[number_c] = 1
                    number_c += 1 
                    ch = 1
    if number_c == number:
        return cost
    var_not_xi = attempt(total_c, number_c + 1, cost)
    if Min < var_not_xi:
        Min = var_not_xi
    return max(var_not_xi, var_xi)

Min = greed(total, number)
print(attempt(total, 0, 0))
```
